```python
#Write a program to draw a scatter plot for random 500 x and y coordinates and style it
import matplotlib.pyplot as plt
import numpy as np
fig = plt.figure()
x = np.random.normal(1, 5, 500)
y = np.random.normal(1, 5, 500)
colour = np.arctan2(x, y)
plt.scatter(x, y, s = 50, c = colour, alpha = 0.8)
plt.colorbar()
plt.show()
```
