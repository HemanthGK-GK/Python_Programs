```python
# Write a program to draw a point plot for sex against survived for a dataset given in ur
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style
sns.set_style('darkgrid')
# Import Data
titanic_dataset = sns.load_dataset("titanic")

# Construct plot
sns.pointplot(x = "sex", y = "survived", data = titanic_dataset)
plt.show()
```
