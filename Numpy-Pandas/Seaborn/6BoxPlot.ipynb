```python
#Write a program to draw box plot of life expectancy by continent for a data set given in a url

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
# Set Seaborn style
sns.set_style('darkgrid')
# Import Data
#tips = sns.load_dataset('Yeardata')
df = pd.read_csv("Yeardata.csv")

# Construct plot
df.boxplot(by ='continent', column=['lifeExp'], grid = True)
plt.show()
```

```python

# Set Seaborn style
sns.set_style('darkgrid')
# Import Data
year_dataset = sns.load_dataset("iris")

# Construct plot
sns.boxplot( x = "species", y = "sepal_length", data = year_dataset)
plt.show()
```
