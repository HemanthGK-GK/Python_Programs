```python
# Write a Python program to plot two or more lines and set the line markers
import matplotlib.pyplot as plt
x = [1,4,5,6,7]
y = [2,6,3,6,3]

# plotting the points 
plt.plot(x, y, color='green', linewidth = 2,marker='o', markerfacecolor='yellow', markersize=7)
plt.ylim(1,9)#Set the y-limits of the current axes.
plt.xlim(1,8)#Set the x-limits of the current axes.

plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('Display marker')# giving a title
plt.show()# function to show the plot
```
