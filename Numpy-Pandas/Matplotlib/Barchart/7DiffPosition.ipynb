```python
# Write a Python programming to display a bar chart of the popularity of programming
# Languages. Specify the position of each bar plot.
# Sample data:
# Programming languages: Java, Python, PHP, JavaScript, C#, C++
# Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7
import matplotlib.pyplot as plt
x = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]
x_pos = [i for i, _ in enumerate(x)]
plt.xlabel("Programming Languages")
plt.ylabel("Popularity")
plt.title("PopularitY of Programming Language")
plt.xticks(x_pos, x)
# Select the position of each barplots on the x-axis (space=1,3,3,2,1)
y_pos = [0,1,3,4,8,10]
# Create bars
plt.bar(y_pos, popularity)
# Turn on the grid
plt.minorticks_on()
plt.grid(which='major', linestyle='-', linewidth='0.5', color='green')
# Customize the minor grid
plt.grid(which='minor', linestyle=':', linewidth='0.5', color='gray')
plt.show()
```
