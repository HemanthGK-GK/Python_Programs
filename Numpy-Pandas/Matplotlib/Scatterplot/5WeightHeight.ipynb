```python
#  Write a Python program to draw a scatter plot for three different groups camparing weights
# and heights.
import matplotlib.pyplot as plt
import numpy as np 
weight1=[10,25,50,64,95]
height1=[16.5,45.3,85.4,71.9,99.0] 
weight2=[75,68,57,62,94]
height2=[15.2,54.5,84.3,61.1,27.6] 
weight3=[73,45,86,51,49]
height3=[65.8,70.9,12.8,35.4,14.6]
weight=np.concatenate((weight1,weight2,weight3))
height=np.concatenate((height1,height2,height3))
plt.scatter(weight, height, color='green')
plt.xlabel('weight', fontsize=16)
plt.ylabel('height', fontsize=16)
plt.title(' Weight vs Height scatter plot',fontsize=20)
plt.show()

```
