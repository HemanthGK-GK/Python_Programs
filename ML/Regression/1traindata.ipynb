```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#preprocessing input data
traindata = pd.read_csv('test.csv')
x=traindata.iloc[:,0]
y=traindata.iloc[:,1]
plt.scatter(x,y)
plt.show()
```

```python
#building model
m=0
c=0
l=0.0001 #lerarning rate
epoch=1000

n=float(len(x))

for i in range(len(x)):
    ypred= m*x + c 
    Dm=(-2/n) * sum(x*(y-ypred))#predict the derivative WRT m
    Dc=(-2/n) * sum(y-ypred) #predict the derivative WRT c
    #updating the derivatives
    m= m - l*Dm
    c= c - l*Dc
print(m,c)
```

```python
#making prediction
ypred= m*x + c

plt.scatter(x,y,c='g')
plt.plot(x,ypred,c='y')
plt.show()
```
